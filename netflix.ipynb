# Univariate Analysis

```python
rating_counts = df['Rating'].value_counts()
```

```python
# Plot Rating Distribution
#plt.figure(figsize=(14, 8))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='viridis')
plt.title('Distribution of Ratings', fontsize=15)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.show()
```

# Bivariate Analysis


```python
# Filter data into TV shows and movies
tv_shows = df[df['Category'] == 'TV Show']
movies = df[df['Category'] == 'Movie']
```

```python
# Group by 'Release_Date' and count number of releases
tv_shows_count = tv_shows.groupby('Release_Date').size().reset_index(name='count')
movies_count = movies.groupby('Release_Date').size().reset_index(name='count')
```

```python
# Plot Release Date vs. Category
#plt.figure(figsize=(14, 8))
plt.plot(tv_shows_count['Release_Date'], tv_shows_count['count'], label='TV Shows', marker='o')
plt.plot(movies_count['Release_Date'], movies_count['count'], label='Movies', marker='o')
plt.title('Release Date vs. Category (TV Shows vs. Movies)', fontsize=15)
plt.xlabel('Release Date', fontsize=12)
plt.ylabel('Number of Releases', fontsize=12)
plt.legend(fontsize=12)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Filter data into TV shows and movies
tv_shows = df[df['Category'] == 'TV Show']
movies = df[df['Category'] == 'Movie']
```

```python
# Count occurrences of TV shows and movies by country
tv_shows_count = tv_shows['Country'].value_counts()
movies_count = movies['Country'].value_counts()
```

```python
# Create a pivot table for TV shows
tv_shows_pivot = tv_shows.pivot_table(index='Country', columns='Category', aggfunc='size', fill_value=0)
```

```python
tv_shows_pivot
```

```python
# Create a pivot table for movies
movies_pivot = movies.pivot_table(index='Country', columns='Category', aggfunc='size', fill_value=0)
```

```python
movies_pivot
```

```python
# Heatmap for TV Shows
plt.subplot(1, 2, 1)
sns.heatmap(tv_shows_pivot,cbar=True)
plt.title('Distribution of TV Shows by Country')
plt.xlabel('Category')
plt.ylabel('Country')
```

```python
# Heatmap for Movies
plt.subplot(1, 2, 2)
sns.heatmap(movies_pivot,cbar=True)
plt.title('Distribution of Movies by Country')
plt.xlabel('Category')
plt.ylabel('')  # Removing ylabel to avoid overlap

#plt.tight_layout()
plt.show()
```

# Duration Analysis


# ##Analyze TV movies

```python
# Filter the dataset to include only movies
movies = df[df['Category'] == 'Movie']
```

```python
# Convert duration to numeric (extracting only the numeric part and converting to float)
movies['Duration'] = movies['Duration'].str.replace(' min', '').astype(float)
```

```python
# Plotting the distribution of movie durations
plt.figure(figsize=(12, 6))
sns.histplot(movies['Duration'], bins=30, kde=True, color='blue', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

# Analyze TV Show Seasons

```python
# Filter the dataset to include only TV shows
tv_shows = df[df['Category'] == 'TV Show']
```

```python
# Count the number of seasons for each TV show
tv_shows['Season_Count'] = tv_shows['Duration'].str.extract('(\d+)').astype(float)
```

```python
# Plotting the count of TV show seasons
#plt.figure(figsize=(10, 6))
sns.countplot(x='Season_Count', data=tv_shows, palette='viridis')
plt.title('Count of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.show()
```

# Top Directors Analysis (Bar Plot)

```python
# Count the number of entries for each director
director_counts = df['Director'].value_counts().sort_values(ascending=False)
```

```python
# Select top 10 directors
top_directors = director_counts.head(10)
```

```python
# Plotting the bar plot
#plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='viridis')
plt.title('Top 10 Directors with Most Content on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```
